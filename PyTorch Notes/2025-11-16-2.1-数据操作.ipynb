## 2.1.5 èŠ‚çœå†…å­˜

è¿è¡Œä¸€äº›æ“ä½œå¯èƒ½ä¼šå¯¼è‡´ä¸ºæ–°ç»“æœåˆ†é…å†…å­˜ã€‚ ä¾‹å¦‚ï¼Œå¦‚æœæˆ‘ä»¬ç”¨Y = X + Yï¼Œæˆ‘ä»¬å°†å–æ¶ˆå¼•ç”¨YæŒ‡å‘çš„å¼ é‡ï¼Œè€Œæ˜¯æŒ‡å‘æ–°åˆ†é…çš„å†…å­˜å¤„çš„å¼ é‡ã€‚

åœ¨ä¸‹é¢çš„ä¾‹å­ä¸­ï¼Œæˆ‘ä»¬ç”¨Pythonçš„id()å‡½æ•°æ¼”ç¤ºäº†è¿™ä¸€ç‚¹ï¼Œ å®ƒç»™æˆ‘ä»¬æä¾›äº†å†…å­˜ä¸­å¼•ç”¨å¯¹è±¡çš„ç¡®åˆ‡åœ°å€ã€‚ è¿è¡ŒY = Y + Xåï¼Œæˆ‘ä»¬ä¼šå‘ç°id(Y)æŒ‡å‘å¦ä¸€ä¸ªä½ç½®ã€‚ è¿™æ˜¯å› ä¸ºPythoné¦–å…ˆè®¡ç®—Y + Xï¼Œä¸ºç»“æœåˆ†é…æ–°çš„å†…å­˜ï¼Œç„¶åä½¿YæŒ‡å‘å†…å­˜ä¸­çš„è¿™ä¸ªæ–°ä½ç½®ã€‚

```python
before = id(Y)
Y = Y + X
id(Y) == before
```

å¹¸è¿çš„æ˜¯ï¼Œæ‰§è¡ŒåŸåœ°æ“ä½œéå¸¸ç®€å•ã€‚ æˆ‘ä»¬å¯ä»¥ä½¿ç”¨åˆ‡ç‰‡è¡¨ç¤ºæ³•å°†æ“ä½œçš„ç»“æœåˆ†é…ç»™å…ˆå‰åˆ†é…çš„æ•°ç»„ã€‚ä¸ºäº†è¯´æ˜è¿™ä¸€ç‚¹ï¼Œæˆ‘ä»¬é¦–å…ˆåˆ›å»ºä¸€ä¸ªæ–°çš„çŸ©é˜µZï¼Œå…¶å½¢çŠ¶ä¸å¦ä¸€ä¸ªYç›¸åŒï¼Œ ä½¿ç”¨zeros_likeæ¥åˆ†é…ä¸€ä¸ªå…¨0çš„å—ã€‚

```python
Z = torch.zeros_like(Y)
print('id(Z):', id(Z))
Z[:] = X + Y
print('id(Z):', id(Z))
```

å¦‚æœåœ¨åç»­è®¡ç®—ä¸­æ²¡æœ‰é‡å¤ä½¿ç”¨Xï¼Œ æˆ‘ä»¬ä¹Ÿå¯ä»¥ä½¿ç”¨X[:] = X + Yæˆ–X += Yæ¥å‡å°‘æ“ä½œçš„å†…å­˜å¼€é”€ã€‚

```python
before = id(X)
X += Y
id(X) == before
```
