## 2.5.1 ä¸€ä¸ªç®€å•çš„ä¾‹å­

å®šä¹‰ä¸€ä¸ªå‡½æ•°y=2XT*Xï¼Œå…³äºåˆ—å‘é‡xæ±‚å¯¼ã€‚é¦–å…ˆï¼Œæˆ‘ä»¬åˆ›å»ºå˜é‡xï¼Œå¹¶ä¸ºå…¶åˆ†é…ä¸€ä¸ªåˆå§‹å€¼ã€‚

```python
x = torch.arange(4.0)
x
```

x.requires_grad_(True) -- è®©å¼ é‡xå¼€å¯æ¢¯åº¦è·Ÿè¸ªï¼ˆgradient trackingï¼‰ã€‚ä¹‹åæ‰€æœ‰å¯¹xçš„è¿ç®—ï¼ˆåŠ ã€å‡ã€ä¹˜ã€ç‚¹ç§¯ã€çŸ©é˜µä¹˜æ³•ã€æ¿€æ´»å‡½æ•°â€¦â€¦ï¼‰ï¼ŒPyTorch éƒ½ä¼šè®°å½•åœ¨è®¡ç®—å›¾ (computational graph) ä¸­ã€‚

```python
x.requires_grad_(True)  # ç­‰ä»·äºx=torch.arange(4.0,ruires_grad=True)
x.grad  # è¿˜æ²¡è°ƒç”¨backward()ä¹‹å‰ï¼Œè¿™é‡Œçš„é»˜è®¤å€¼æ˜¯None
```

å‡½æ•°y=2XT*Xï¼Œå…³äºxçš„æ¢¯åº¦åº”è¯¥ä¸º4xï¼ŒéªŒè¯ä¸€ä¸‹è¿™ä¸ªæ¢¯åº¦æ˜¯å¦è®¡ç®—æ­£ç¡®

```python
y = 2 * torch.dot(x, x)
y
```

è°ƒç”¨åå‘ä¼ æ’­å‡½æ•°æ¥è‡ªåŠ¨è®¡ç®—yå…³äºxæ¯ä¸ªåˆ†é‡çš„æ¢¯åº¦ï¼Œå¹¶æ‰“å°è¿™äº›æ¢¯åº¦ã€‚

```python
y.backward()
x.grad
```

```python
x.grad == 4 * x
```

```python
# åœ¨é»˜è®¤æƒ…å†µä¸‹ï¼ŒPyTorchä¼šç´¯ç§¯æ¢¯åº¦ï¼Œæˆ‘ä»¬éœ€è¦æ¸…é™¤ä¹‹å‰çš„å€¼
x.grad.zero_()
y = x.sum()
y.backward()
x.grad
```
